```python
clip_len = 32
num_clips=1
frame_interval = 1

ori_clip_len = 32 * 1 
num_frames = 400

avg_interval = (300-32+1) // 32
avg_interval
```

```python
base_offsets = np.arange(num_clips) * avg_interval
clip_offsets = base_offsets + np.random.randint(avg_interval, size=num_clips)
clip_offsets
```

```python
ratio = (num_frames - ori_clip_len + 1.0) / num_clips
clip_offsets = np.around(np.arange(num_clips) * ratio)
clip_offsets
```

```python
total_frames =30
frame_inds = clip_offsets[:, None] + np.arange(clip_len)[None, :] * frame_interval
frame_inds = np.concatenate(frame_inds)
frame_inds = frame_inds.reshape((-1, clip_len))
frame_inds = np.mod(frame_inds, total_frames)

start_index = 0
frame_inds = np.concatenate(frame_inds) + start_index
frame_inds
```

```python
import tensorflow as tf
```

```python
video = tf.random.uniform((3,32,256,256,3), 0,255, tf.dtypes.int32)
```

```python
result = tf.image.random_crop(video, (3,32,224,224,3))
result.shape
```

```python
import random
images_list = tf.unstack(result)
for i in range(3):
    
    isFlip = random.choice(["flip", "don't flip"])
    if isFlip == "flip":
        images_list[i] = tf.image.flip_left_right(images_list[i])

result = tf.stack(images_list)

result.shape
```

```python
tf.image.per_image_standardization(
    result
)
```
